- Use the linregress function from scipy.stats to get the slope and y-intercept of the line of best fit. Plot the line of best fit over the top of the scatter plot. Make the line go through the year 2050 to predict the sea level rise in 2050.
- Plot a new line of best fit just using the data from year 2000 through the most recent year in the dataset. Make the line also go through the year 2050 to predict the sea level rise in 2050 if the rate of rise continues as it has since the year 2000.

<h1 align="center">Setup</h1>

## Step 1: Import Libraries

Import and configure libraries required for data analysis.

```python
import numpy as np
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.stats import linregress
```

## Step 2: Import Dataset
Import dataset as variable `df` into Python.

```python
url = "https://raw.githubusercontent.com/blackcrowX/Data-Analysis-Projects/main/Python/Sea_Level_Predictor/epa_sea_level.csv"
df = pd.read_csv(url)
```

<h1 align="center">Data Visualisation</h1>

## Step 3: Create Scatter Plot

xyz

```python
# Create scatter plot
f, ax = plt.subplots(1,1, figsize = (10,5))
ax = plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'])
```

## Step 4: Create first line of best fit

xyz

```python
f, ax = plt.subplots(1,1, figsize = (10,5))
lineA = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
xA = np.arange(df['Year'].min(),2050,1)
yA = xA*lineA.slope + lineA.intercept
ax = plt.plot(xA,yA)
```

## Step 5: Create second line of best fit

xyz and add labels and title.

```python
f, ax = plt.subplots(1,1, figsize = (10,5))
df_2000 = df[df['Year'] >= 2000]
lineB = linregress(df_2000['Year'], df_2000['CSIRO Adjusted Sea Level'])
xB = np.arange(2000,2050,1)
yB = xB*lineB.slope + lineB.intercept

plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')

ax = plt.plot(xB,yB)
```
